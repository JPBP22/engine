```python
'''
This function formats the game name to remove special characters, spaces and lowercases all letters
'''

def formatter(game):
    
    game = game.replace("'","").replace("-"," ").replace(",","").replace(".","").lower()
    
    words = game.split()

    return "-".join(words)
# Combine all the data, both games and game details
# Importing the necessary libraries and packages
import aiohttp
import asyncio
import pandas as pd
import requests
import csv
from bs4 import BeautifulSoup
import os
import time
import random
import os
import time
import random

def nordvpn():
    windows_countries = ['United States', 'Canada', 'Argentina', 'Brazil', 'Mexico', 'Costa Rica', 'Chile', 
                        'United Kingdom', 'Germany', 'France', 'Netherlands', 'Sweden', 'Switzerland', 
                        'Denmark', 'Poland', 'Italy', 'Spain', 'Norway', 'Belgium', 'Ireland', 'Czech Republic', 
                        'Austria', 'Portugal', 'Finland', 'Ukraine', 'Romania', 'Serbia', 'Hungary', 'Luxembourg', 
                        'Slovakia', 'Bulgaria', 'Latvia', 'Greece', 'Iceland', 'Estonia', 'Albania', 'Croatia', 
                        'Cyprus', 'Slovenia', 'Moldova', 'Bosnia and Herzegovina', 'Georgia', 'North Macedonia', 
                        'Turkey', 'South Africa', 'India', 'Israel', 'Turkey', 'United Arab Emirates', 'Australia', 
                        'Taiwan', 'Singapore', 'Japan', 'Hong Kong', 'New Zealand', 'Malaysia', 'Vietnam', 'Indonesia', 
                        'South Korea', 'Thailand']
    os.chdir(r"C:\Program Files\NordVPN")
    country = random.choice(windows_countries)
    server = f"nordvpn -c -g {country}"
    os.system(server)

    time.sleep(8)

#Async function that retrieves the game information by making a call to the GiantBomb API

async def retrieve_games_info(game_name, game_id, api_key = "69fe0db0bb287eeb835b3edcbab1388545aeb099"):
    # API endpoint to retrieve game details
    endpoint = f"https://store.steampowered.com/api/appdetails?appids={game_id}"

    # API key, you'll need to sign up for a free API key to use the GiantBomb API

    # Query parameters for the API request

    headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3",
    "Accept": "*/*",
    "Referer": "https://store.steampowered.com",
    "Connection": "keep-alive"
    }

    async with aiohttp.ClientSession(connector=aiohttp.TCPConnector(ssl=False)) as session:
        async with session.get(endpoint, headers=headers) as response:
            # Check if the response was successful
            if response.status == 200:
                # Extract the games information from the response
                response_json = await response.json()

                #if response_json["status_code"] == 107:
                #    return 107
                k = list(response_json.keys())[0]
                if response_json[k]['success'] == False:
                    return 107
                response_data = response_json[k]['data']
                if response_data['type'] == "game":
                    return response_data
            else:
                raise Exception(f"Failed to retrieve games info. Status code: {response.status}")
# Async function that fetches the game details from Steam by making a call to the Steam API
async def fetch_game_details(s,appid):
    async with aiohttp.ClientSession(connector=aiohttp.TCPConnector(ssl=False)) as session:
        async with session.get(f"https://steamspy.com/api.php?request=appdetails&appid={appid}") as resp:
            return await resp.json()
# Async function that fetches all games by making a call to the Steam API
async def fetch_games(session):
    async with session.get("https://api.steampowered.com/ISteamApps/GetAppList/v2") as resp:
        return await resp.json()
# Async function that saves the game details to a CSV file
import csv

async def save_games_to_csv(session, games, fromGame=0, toGame=1000):
    c = fromGame
    key_index = 0
    with open("all_games.csv", "a", newline="", encoding="utf-8") as csvfile:
        writer = csv.writer(csvfile)

        for game in games["applist"]["apps"][fromGame:toGame]:
            print(c)
            c += 1
            game_details = await fetch_game_details(session, game["appid"])
            try:
                further_details = await retrieve_games_info(formatter(game['name']), game['appid'])
                print(game["name"])
                if further_details == 107:
                    print('Game not found')
                    continue

            except Exception as e:
                print("using nordvpn")
                nordvpn()
                further_details = await retrieve_games_info(formatter(game['name']), game['appid'])
                print(game["name"])
                if further_details == 107:
                    print('Game not found')
                    continue

            if game_details:
                data = game_details
                if further_details:
                    print("retrieving")
                    desc_w_tags = further_details.get("detailed_description", "")
                    release_date = further_details.get("release_date", {})
                    if desc_w_tags:
                        description_text = BeautifulSoup(desc_w_tags, "html.parser").get_text()
                    else:
                        description_text = ""

                    if release_date:
                        if release_date["coming_soon"]:
                            release_date = 'coming soon'
                        else:
                            release_date = release_date.get("date", "")
                    else:
                        release_date = ""
                    writer.writerow([
                        game["appid"], game["name"],
                        data.get("price", "0"), data.get("developer", ""), data.get("publisher", ""),
                        data.get("average_forever", 0), data.get("median_forever", 0), data.get("ccu", 0),
                        data.get("genre", []), data.get("tags", []), data.get("languages", []), data.get("positive", 0),
                        data.get("negative", 0),
                        further_details.get("short_description", ""),
                        description_text,
                        release_date,
                    ])
            else:
                continue



    #df = pd.DataFrame(games_data, columns=["appid", "name", "price", "developer", "publisher", "average_forever", "median_forever", "ccu", "genre", "tags", "languages" ,"positive", "negative", "short_description", "description", "release_date"])
    #df.to_csv("all_games.csv", index=False)

async def main():
    async with aiohttp.ClientSession() as session:
        games = await fetch_games(session)
        await save_games_to_csv(session, games, fromGame=0, toGame=2000)

if __name__ == "__main__":
    await main()
```
